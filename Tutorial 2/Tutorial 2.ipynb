```python
#calculate the tendencies for the bike ages 

bike_age = [12, 13, 17, 2, 23, 15, 27, 5, 11, 8, 13, 5, 20, 14, 3, 17, 8, 10, 21, 16, 13]
print(len(bike_age))

import numpy as np 

mean = np.mean(bike_age)
median  = np.median(bike_age)
first_quantile = np.quantile(bike_age,0.25)
third_quantile = np.quantile(bike_age,0.75)

print('The mean of the bike age is', mean)
print('The median of the bike age is', median)
print('The first quantile of the bike age is', first_quantile)
print('The third quantile of the bike age is', third_quantile)
```

```python
#calculate the tendencies for the bike ages 

bike_age = [12, 13, 17, 2, 23, 15, 27, 5, 11, 8, 13, 5, 20, 14, 3, 17, 8, 10, 21, 16, 13]

import statistics as st 

mean = st.mean(bike_age)
median  = st.median(bike_age)
first_quantile = st.quantiles(bike_age)
third_quantile = st.quantiles(bike_age)

print('The mean of the bike age is', mean)
print('The median of the bike age is', median)
print('The first quantile of the bike age is', first_quantile[0])
print('The third quantile of the bike age is', third_quantile[2])
```

```python
#new vintage bike added to the colection, calculate new central tendencies 

bike_age.extend([255])
print(bike_age)
```

```python
new_mean = np.mean(bike_age)
new_median  = np.median(bike_age)

print('The new mean of the bike age is', new_mean)
print('The new median of the bike age is', new_median)
```
